## **2. Dynamic Prompt Engineering with Templates**

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

print(template.format(topic="LangChain"))
```

# **LLM Initialization (Setup)**

```python
from langchain_openai import ChatOpenAI
import os

# os.environ["OPENAI_API_KEY"] = "your-api-key"

llm = ChatOpenAI(model="gpt-3.5-turbo")
```

## **3. Building Execution Pipelines with Chains**

```python
from langchain_core.runnables import RunnableSequence

chain = template | llm
```

## **5. Tool Integration and Agent Capabilities**

```python
from langchain_core.tools import Tool

def calculator(input):
    return eval(input)

tool = Tool(
    name="Calculator",
    func=calculator,
    description="Performs basic math calculations"
)

print(tool.invoke("5+3"))
```
